```python
heart_df.dropna(inplace = True)
heart_df.info()
```

```python
heart_df.head(10)
```

```python
heart_df['thal'].replace({'fixed defect':'fixed_defect' , 'reversable defect': 'reversable_defect' }, inplace =True)
heart_df['cp'].replace({'typical angina':'typical_angina', 'atypical angina': 'atypical_angina' }, inplace =True)
```

```python
heart_df_temp = heart_df[['age','sex','cp', 'trestbps', 'chol', 'fbs',  'thalch', 'exang', 'oldpeak', 'slope', 'ca', 'thal']].copy()
heart_df_temp['target'] = ((heart_df['num'] > 0)*1).copy()
heart_df_temp['sex'] = (heart_df['sex'] == 'Male')*1
heart_df_temp['fbs'] = (heart_df['fbs'])*1
heart_df_temp['exang'] = (heart_df['exang'])*1
```

```python
heart_df_temp.columns = ['age', 'sex', 'chest_pain_type', 'resting_blood_pressure', 
              'cholesterol', 'fasting_blood_sugar',
              'max_heart_rate_achieved', 'exercise_induced_angina', 
              'st_depression', 'st_slope_type', 'num_major_vessels', 
              'thalassemia_type', 'target']
# heart_df_temp.head(10)
```

```python
heart_df = pd.get_dummies(heart_df_temp, drop_first=False)
heart_df.columns
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X = heart_df.drop('target', axis=1)
y = heart_df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=41)
```

```python
X_train=(X_train-np.min(X_train))/(np.max(X_train)-np.min(X_train)).values
X_test=(X_test-np.min(X_test))/(np.max(X_test)-np.min(X_test)).values
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('The Accuracy Score is: ', accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
```

```python
import shap
```

```python
explainer = shap.Explainer(reg.predict, X_test)
```

```python
shap_values = explainer(X_test)
```

## Feature Importance

```python
shap_df = pd.DataFrame(shap_values.values, columns=shap_values.feature_names, index=X_test.index)
shap_df
```

```python
# calculation for feature importance
```

```python
num = np.abs(shap_df).mean(0)
f_importance = pd.DataFrame(list(zip(X_test.columns, num)), columns=['feature_names', 'feature_importance_values'])
f_importance.sort_values(by=['feature_importance_values'], ascending=False, inplace=True)
f_importance
```
